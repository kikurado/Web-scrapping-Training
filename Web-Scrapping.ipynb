# Web Scrapping for Wuzzuf.com Website.

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
url ='https://wuzzuf.net/search/jobs/?q=Data+analyst+&a=hpb' #getting the Url
```

```python
client = urlopen(url) #connecting to the Url
```

```python
html = client.read() #connection established with the website
```

```python
client.close()# connection ended with the website after obtaining a compy from the Html!
```

```python
soup =  bs (html,'html.parser') #using beauitful soup to organize the data
```

```python
soup # so far so good!
```

```python
#selecting containers in the page to extract the neccessary Data from it
containers = soup.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
```

```python
len(containers) #check the number of the containers
```

```python
#extracting the title from the first container

title = containers[0].findAll('h2',{'class':'css-m604qf'})
```

```python
title[0].text #checking if it's working! yep! all good.
```

```python
#extracting the company's Name from the first container

cname = containers[0].findAll('a',{'class':'css-17s97q8'})
```

```python
cname[0].text.replace(' -',"")
```

```python
#extracting the type of the jobs offered

jtype = containers[0].findAll('div',{'class':'css-1lh32fc'})
```

```python
jtype[0].text
```

```python
#creating a CSV file to contain the result  
f = open(r'C:\Users\woo\Desktop\Jupyter\Web-Scrapping/Wuzzuf.csv','w')
header = 'job_title, company_name, job_type\n' # adding headers to the file!
f.write(header)
```

```python
#extracting all containers with the same method

for container in containers:
    title = container.findAll('h2',{'class':'css-m604qf'})
    job_title =title[0].text.strip()
    
    cname = container.findAll('a',{'class':'css-17s97q8'})
    company_name = cname[0].text.replace(' -',"").strip()
    
    jtype = container.findAll('div',{'class':'css-1lh32fc'})
    job_type = jtype[0].text.strip()
    
    f.write(job_title + ',' + company_name +',' + job_type + '\n')
f.close()    
```
